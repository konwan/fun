# get few articles from ptt NBA

```python
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import jieba

class Ptt(object):
    def __init__(self):
        # Gossiping 20089; NBA 4458
        self.board = "NBA"
        self.idx = "5746"
        self.fp = "/Users/cindy/Desktop/pg/mit/fun"
        self.data = []
    
    def getposts(self):
        driver = webdriver.Chrome("/Users/cindy/Desktop/pg/driver/chromedriver")
        url = 'https://www.ptt.cc/bbs/{}/index{}.html'.format(self.board, self.idx)
        driver.get(url)
        res = BeautifulSoup(driver.page_source, "lxml")
        
        # over 18 check
        btns18 = res.find_all('div', class_="over18-button-container")
        if len(btns18) != 0:
            driver.find_element_by_name("yes").click()
            res = BeautifulSoup(driver.page_source, "lxml")
        
        # get posts
        posts = res.find_all("div", class_ = "r-ent")
        tmp = []
        for post in posts:
            athr = post.find("div", class_ = "author").getText()
            if athr != '-':
                title = post.find('a').text.strip()
                tmpcmt = post.find("div", class_ = "nrec").text
                cmt = int(tmpcmt) if tmpcmt.isdigit() else np.nan
                url = post.find('a')['href']
            else:
                title = "DELETE"
                cmt = 0
                url = ""
            dt = post.find("div", class_ = "date").text
            data = [cmt, dt, athr, title]
            self.data.append(data)
            # driver.save_screenshot("{}/{}.png".format(self.fp, self.idx))
        
        # get next page
        res.find('div', class_ = "btn-group btn-group-paging").find_all('a')[0].getText()
        btns = {}
        for i in res.find_all('a', class_ = "btn wide"):
            k = i.getText().replace("‹","").replace("›","").strip()
            btns[k] = i['href']
        driver.quit() 
        if "下頁" not in btns.keys():
            print("this is the last page {}".format(self.idx))
        elif "上頁" not in btns.keys():
            print("this is the 1st page")
        else: 
            preidx = btns['上頁'].split('/')[3].replace('index','').replace('.html','')
            nextidx = btns['下頁'].split('/')[3].replace('index','').replace('.html','')
            self.idx=nextidx
            self.getposts()

# res.select('#main-container > div.r-list-container.bbs-screen > div:nth-of-type(1) > div.nrec > span')
# res.select('#main-container > div.r-list-container.bbs-screen > div:nth-of-type(1) > div.title > a')        
p = Ptt()
p.getposts()
col = ['recommends','date','author','title']
result = pd.DataFrame(p.data, columns=col)
# result = {"a": lista, "b": listb, "c": listc, "d": listd }
result.to_csv(sep=',', columns=['idx']+col, path_or_buf="{}/{}_{}.txt".format(p.fp, p.board, p.idx))
```
